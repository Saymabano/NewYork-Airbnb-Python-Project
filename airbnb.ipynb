Data Cleaning

```python
data.isnull().sum()

# dropping null values
data.dropna(inplace=True)

data.isnull().sum()
```

```python
data.shape
```

```python
data.duplicated().sum()
[data.duplicated()]

#delete duplicate rows

data.drop_duplicates(inplace=True)
```

```python
data.shape
```

```python
data.dtypes
```

```python
# type casting

data['id'] = data['id'].astype(object)

data['host_id'] = data['host_id'].astype(object)

data.dtypes
```

Data Analysis

```python
data['price']
```

```python
# identifying outliers
```

```python
df = data[data['price'] < 1500]

sns.boxplot(data=df, x='price')
```

```python
# price distribution
plt.figure(figsize=(15,5))
plt.title('Price Distribution')
sns.histplot(data = df,x='price')
plt.ylabel('Frequency')
```

```python
# price distribution
plt.figure(figsize=(5,3))
plt.title('Availability Distribution')
sns.histplot(data = df, x='availability_365')
plt.ylabel('Frequency')
```

**Feature Engineering**

```python
# price per bed

df['price per bed'] = df['price']/df['beds']
```

```python
df.head()
```

```python
# price dependency on neighbourhood
sns.barplot(data=df ,x='neighbourhood_group', y='price', hue='room_type')
```

<!-- number of reviews and price  -->

```python
plt.figure(figsize=(8,5))
plt.title('Locality and Review dependecy')
sns.scatterplot(data=df , x='number_of_reviews',y='price',hue='neighbourhood_group')
plt.show()
```
